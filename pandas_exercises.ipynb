# Task 16: 
Write a function that returns 1 for those under 30 and 0 for those above or equal to 30.
 
Using the function you wrote, create a variable named 'age_flag' in the titanic dataset (use apply and lambda constructs).

```python
# Function, apply & lambda
def age(a):
    if a < 30:
        return 1
    else:
        return 0

df['age_flag'] = df['age'].apply(lambda x: age(x))
df['age_flag'].head()
```

```python
# Only with apply & lambda
df['age_flag'] = df['age'].apply(lambda x: 1 if x < 30 else 0)
df['age_flag'].head()
```
